# NN using Numpy

```python
import numpy as np
```

```python
X = np.array([7,2,25])
actual = 85
```

```python
# weights and biases (taken randomly)

w1, w2, w3 = 0.1, 0.1, 0.1
w4, w5, w6 = 0.16, 0.5, 0.6
w7, w8 = 7, 3
b1, b2, b3 = -4, -4, 12
```

```python
print(f"w1: {w1}")
print(f"w2: {w2}")
print(f"w3: {w3}")
print(f"w4: {w4}")
print(f"w5: {w5}")
print(f"w6: {w6}")
```

```python
def sigmoid(x):
  return 1/ (1 + np.exp(-x))
```

```python
# Forward Propagation
```

```python
def forward_progagation(X):
  z1 = (w1*X[0]) + (w2*X[1]) +(w3*X[2]) +b1
  z2 = (w3*X[0]) + (w4*X[1]) + (w5*X[2]) +b2
  g1 = sigmoid(z1)
  g2 = sigmoid(z2)
  y_pred = (g1*w7) + (g2*w8) + b3
  return z1, z2, g1, g2, y_pred
```

```python
z1, z2, g1, g2, y_pred = forward_progagation(X)
print(f'z1: {z1}')
print(f'z2: {z2}')
print(f'g1: {g1}')
print(f'g2: {g2}')
print(f'y_pred: {y_pred}')
print(f'cost: {actual - y_pred}')
```

```python
# Backward Propagation

def sigmoid_derivative(x):
  return x * (1-x)
```

```python
def backward_propagation(X, actual, z1, z2, g1, g2, y_pred):
  dc_dy_pred = 2 * (y_pred - actual)

  dc_dw7 = dc_dy_pred * g1
  dc_dw8 = dc_dy_pred * g2
  dc_db3 = dc_dy_pred * 1

  dc_dg1 = dc_dy_pred * w7
  dc_dg2 = dc_dy_pred * w8

  dc_dz1 = dc_dg1 * sigmoid_derivative(g1)
  dc_dz2 = dc_dg2 * sigmoid_derivative(g2)

  dc_dw1 = dc_dz1 * X[0]
  dc_dw2 = dc_dz1 * X[1]
  dc_dw3 = dc_dz1 * X[2]
  dc_dw4 = dc_dz2 * X[0]
  dc_dw5 = dc_dz2 * X[1]
  dc_dw6 = dc_dz2 * X[2]
  dc_db1 = dc_dz1 * 1
  dc_db2 = dc_dz2 * 1

  return dc_dw1, dc_dw2, dc_dw3, dc_dw4, dc_dw5, dc_dw6, dc_dw7, dc_dw8, dc_db1, dc_db2, dc_db3

```

```python
gradients = backward_propagation( X, actual, z1, z2, g1, g2, y_pred)
```

```python
print(f'dc/dw1 = {gradients[0]}')
print(f'dc/dw2 = {gradients[1]}')
print(f'dc/dw3 = {gradients[2]}')
print(f'dc/dw4 = {gradients[3]}')
print(f'dc/dw5 = {gradients[4]}')
print(f'dc/dw6 = {gradients[5]}')
print(f'dc/dw7 = {gradients[6]}')
print(f'dc/dw8 = {gradients[7]}')
print(f'dc/db1 = {gradients[8]}')
print(f'dc/db2 = {gradients[9]}')
print(f'dc/db3 = {gradients[10]}')
```

```python
# update weights

w1 = w1 - 0.001 * gradients[0]
w2 = w2 - 0.001 * gradients[1]
w3 = w3 - 0.001 * gradients[2]
w4 = w4 - 0.001 * gradients[3]
w5 = w5 - 0.001 * gradients[4]
w6 = w6 - 0.001 * gradients[5]
w7 = w7 - 0.001 * gradients[6]
w8 = w8 - 0.001 * gradients[7]
b1 = b1 - 0.001 * gradients[8]
b2 = b2 - 0.001 * gradients[9]
b3 = b3 - 0.001 * gradients[10]

```

```python
print(f"Updated w1 = {w1}")
print(f"Updated w2 = {w2}")
print(f"Updated w3 = {w3}")
print(f"Updated w4 = {w4}")
print(f"Updated w5 = {w5}")
print(f"Updated w6 = {w6}")
print(f"Updated w7 = {w7}")
print(f"Updated w8 = {w8}")
print(f"Updated b1 = {b1}")
print(f"Updated b2 = {b2}")
print(f"Updated b3 = {b3}")
```

# Model Building - from scratch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow import keras
```

```python
# creating our own sythetic data
np.random.seed(10)
num_samples = 300
X = np.random.rand(num_samples, 3) * np.array([24, 10 , 60]) # 24 hours, 10 cups of coffee, 60 minutes travel time
y = np.random.rand(num_samples) * 100
```

```python
#When you set a seed, NumPy starts generating random numbers in a deterministic way.
 #That means if you run the same code multiple times after setting the same seed,
 #youвҖҷll get the exact same random numbers each time.
```

```python
pd.DataFrame(X, columns=['hours_of_sleep', 'cups_of_coffee', 'travel_time']).head()
```

```python
len(X)
```

```python
# NEURAL NETWORK

def create_custom_model():

  #input layer
  inputs = keras.Input(shape=(3,)) #3 input nodes/neurons

  #hidden layer
  x = layers.Dense(2, activation='sigmoid')(inputs)
  # Dense: each input node is connected with both of the hidden nodes
  # (inputs):the connection between input and hidden layer is made

  #output layer
  outputs = layers.Dense(1, activation='linear')(x)
  # activation = 'linear': since no activation is used from hidden to output (linear means no activation function)
  # (x): the connection between hidden and output layer is made

  model = keras.Model(inputs = inputs, outputs = outputs)
  # we are calling the whole layers as a model
  # i.e. calling a function with a variable called model

  model.compile(optimizer = keras.optimizers.Adam(learning_rate=0.01),
                loss = 'mean_squared_error',
                metrics = ['accuracy'])
  # вҖңLetвҖҷs move fast in consistent directions (momentum), but
  #   also adapt how far we step for each parameter (adaptive learning rates).вҖқ

  return model
```

```python
model = create_custom_model()
```

```python
model.summary()
```

```python
history = model.fit(X,y, epochs = 130,
                    batch_size=32, validation_split=0.20)
```

```python
plt.figure(figsize=(12,4))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

### Feature Selection Technique

```python
df.head()
```

```python
# converting categorical features into numeric feacutres

numerical_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'tenure', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'MonthlyCharges', 'TotalCharges',]

for col in numerical_features:
  if df[col].dtype == 'object':
    df[col] = pd.factorize(df[col])[0]
    # [0] we are just taking the first output of the above line([0] is indexing)
# combining all

numerical_features = [f for f in numerical_features if f in df.columns]
# Ensures all the features are included
```

#### Scaling

```python
X = df[selected_features]
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=10)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


with open('scaler.pkl', 'wb') as f:   # wb - write binary
  pickle.dump(scaler, f)

```

#### Initial Model

```python
def create_model(hp=None):
    if hp:
        model = models.Sequential()
        model.add(layers.Dense(
            units=hp.Choice('units_1', values=[32, 64, 128]),
            activation='relu',
            input_shape=(X_train_scaled.shape[1],)
        ))
        if hp.Boolean('second_layer'):
            model.add(layers.Dense(
                units=hp.Choice('units_2', values=[16, 32]),
                activation='relu'
            ))
        model.add(layers.Dense(1, activation='sigmoid'))
        model.compile(
            optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy']
        )
    else:
      #5 ---> hl1(64) ----> hl2(32) ----> hl3(16) ----> hl4(8) ----->1
        model = models.Sequential([
            layers.Dense(64, activation='relu',
                         input_shape=(X_train_scaled.shape[1],)),
            layers.Dense(32, activation='relu'),
            layers.Dense(16, activation='relu'),
            layers.Dense(8, activation='relu'),
            layers.Dense(1, activation='sigmoid')
        ])
        model.compile(optimizer="adam", loss='binary_crossentropy',
                      metrics=['accuracy'])

    return model

initial_model = create_model()
initial_model.summary()

history = initial_model.fit(
    X_train_scaled, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.20,
    verbose=1
)
```

### Predictions

```python
# predictions

def load_model_components():
  model = tf.keras.models.load_model('best_churn_model.h5')
  with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
  with open('selected_features.pkl', 'rb') as f:
    selected_features = pickle.load(f)
  return model, scaler, selected_features

def predict_churn(*args):
  model, scaler, selected_features = load_model_components()
  input_data = pd.DataFrame([args], columns=selected_features)
  input_scaled = scaler.transform(input_data)
  prediction = model.predict(input_scaled)[0][0]
  probability = float(prediction)
  churn_status = 'Yes' if probability > 0.5 else 'No'

  return{
      'Churn Probability': probability,
      'Churn Status': churn_status
  }
```
